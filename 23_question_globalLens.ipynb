```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("drive/MyDrive/dades.csv")
df = df[2:] #Les dos primeres files son NaN
```

```python
q_df = df.loc[:,"i1":"i69"] #dataframe amb totes les preguntes
```

BENESTAR EMOCIONAL (i1-i8)

```python
be_df = df.loc[:,"i1":"i8"]
df["EW"] = be_df.sum(axis=1)

ri_df = df.loc[2:, "i9":"i18"]
df["IR"] = ri_df.sum(axis=1)

bm_df = df.loc[2:, "i19":"i26"]
df["MW"] = bm_df.sum(axis=1)

dp_df = df.loc[2:, "i27":"i34"]
df["PD"] = dp_df.sum(axis=1)

bf_df = df.loc[2:, "i35":"i42"]
df["PW"] = bf_df.sum(axis=1)

ad_df = df.loc[2:, "i43":"i51"]
df["SD"] = ad_df.sum(axis=1)

is_df = df.loc[2:, "i52":"i59"]
df["SI"] = is_df.sum(axis=1)

dr_df = df.loc[2:, "i60":"i69"]
df["RI"] = dr_df.sum(axis=1)

df_dimensions = df.loc[:, "EW":"RI"]
df_dimensions["QOL"] = df_dimensions.sum(axis=1)
```

```python
barem_df = pd.read_csv("drive/MyDrive/barem.csv")
barem_df
```

```python
df_dimensions_s = df.loc[:, "EW":"RI"]

# Función para obtener la puntuación correspondiente
def get_standard_value(column, value):
    for i, row in barem_df.iterrows():
      if pd.isna(row[column]):
        continue
      if '-' in row[column]:  # Si es un rango
        low, high = map(int, row[column].split('-'))
        if low <= value <= high:
            return row['Score']
      elif row[column] != '':  # Si está vacío
        if value == int(row[column]):
          return row['Score']
    return np.nan

# Aplicar la función a cada valor en df
for column in df_dimensions_s.columns:
    print(column)
    df_dimensions_s[column] = df_dimensions_s[column].apply(lambda x: get_standard_value(column, x))

df_dimensions_s
```

```python
df_dimensions_s['QOL_s']= df_dimensions_s.sum(axis=1)
```

```python
df_dimensions_s["EW_n"] = df_dimensions_s['EW'] / 20
df_dimensions_s["IR_n"] = df_dimensions_s['IR'] /20
df_dimensions_s["MW_n"] = df_dimensions_s['MW'] /20
df_dimensions_s["PD_n"] = df_dimensions_s['PD'] /20
df_dimensions_s["PW_n"] = df_dimensions_s['PW'] /20
df_dimensions_s["SD_n"] = df_dimensions_s['SD'] /20
df_dimensions_s["SI_n"] = df_dimensions_s['SI'] / 20
df_dimensions_s["RI_n"] = df_dimensions_s['RI']/20
df_dimensions_s["QOL_sn"] = df_dimensions_s['QOL_s'] /160
df_dimensions_s_normalized =  df_dimensions_s.loc[:, "EW_n":"QOL_sn"]
```

```python
for column in df_dimensions_s.loc[:, "EW":"RI"].columns:
  df_dimensions_s[column + '_mod'] = np.select(
    [df_dimensions_s[column] < 7,
     df_dimensions_s[column].between(7, 10),
     df_dimensions_s[column].between(11, 13),
     df_dimensions_s[column].between(13, 20)],
    [1, 0, 0, 1],
    np.nan)
  df_dimensions_s[column + '_level'] = np.select(
    [df_dimensions_s[column] < 7,
     df_dimensions_s[column].between(7, 10),
     df_dimensions_s[column].between(11, 13),
     df_dimensions_s[column].between(13, 20)],
    [0, 0, 1, 1],
    np.nan)

df_dimensions_s['QOL_label'] = np.select(
    [df_dimensions_s["QOL_s"] > 115,
     df_dimensions_s["QOL_s"] < 86,
     df_dimensions_s["QOL_s"].between(86, 95),
     df_dimensions_s["QOL_s"].between(96, 105),
     df_dimensions_s["QOL_s"].between(106, 115)],
    ['very-high', 'very-low', 'low', 'medium', 'high'],
    np.nan
)

df_dimensions_s['QOL_labeln'] = np.select(
    [df_dimensions_s["QOL_s"] > 115,
     df_dimensions_s["QOL_s"] < 86,
     df_dimensions_s["QOL_s"].between(86, 95),
     df_dimensions_s["QOL_s"].between(96, 105),
     df_dimensions_s["QOL_s"].between(106, 115)],
    #[4, 0, 1, 2, 3],
    [0.9, 0.1, 0.3, 0.5, 0.7],
    np.nan
)
df_dimensions_s
```

PREPare dataset to work with LENs

```python
df_LENs = df_dimensions_s.loc[:, "EW":"RI"]
df_LENs['QOL'] = df_dimensions_s['QOL_s']

# Función para obtener la puntuación correspondiente
def get_standard_value(column, value):
    for i, row in barem_df.iterrows():
      if pd.isna(row[column]):
        continue
      if '-' in row[column]:  # Si es un rango
        low, high = map(int, row[column].split('-'))
        if low <= value <= high:
            return row['Score']
      elif row[column] != '':  # Si está vacío
        if value == int(row[column]):
          return row['Score']
    return np.nan

# Aplicar la función a cada valor en df
for column in df_LENs.columns:
    print(column)
    df_LENs[column + '_vlow'] = np.zeros(len(q_df))
    df_LENs[column + '_low'] = np.zeros(len(q_df))
    df_LENs[column + '_med'] = np.zeros(len(q_df))
    df_LENs[column + '_high'] = np.zeros(len(q_df))
    df_LENs[column + '_vhigh'] = np.zeros(len(q_df))
```

```python
# Aplicar la función a cada valor en df
for column in df.loc[:, "EW":"RI"].columns:
  print(column)
  for i, row in df_LENs.iterrows():
    if(row[column] <= 6):
      df_LENs.loc[i, column + '_vlow'] = 1
    elif(row[column] <= 9 ):
      df_LENs.loc[i, column + '_low'] = 1
    elif(row[column] <= 11 ):
      df_LENs.loc[i, column + '_med'] = 1
    elif(row[column] <= 13 ):
      df_LENs.loc[i, column + '_high'] = 1
    elif(row[column] <= 20 ):
      df_LENs.loc[i, column + '_vhigh'] = 1

column = "QOL"
for i, row in df_LENs.iterrows():
  if(row[column] < 76):
    df_LENs.loc[i, column + '_vlow'] = 1
  elif(row[column] < 86 ):
    df_LENs.loc[i, column + '_low'] = 1
  elif(row[column] < 96 ):
    df_LENs.loc[i, column + '_med'] = 1
  elif(row[column] < 106 ):
    df_LENs.loc[i, column + '_high'] = 1
  elif(row[column] < 138 ):
    df_LENs.loc[i, column + '_vhigh'] = 1
```

Only one class explanations for the 23 question architecture

```python
import torch
import torch_explain as te
from sklearn.model_selection import train_test_split
from torch_explain.nn.functional import prune_equal_fanin
from torch_explain.logic.nn import entropy
from sklearn.metrics import accuracy_score
```

Very-low

```python
columns = [3, 9, 11, 13, 15, 20, 21, 24, 27, 30, 36, 37, 38, 43, 45, 46, 49, 50, 52, 53, 56, 58, 59]
x_train, x_test, c_train, c_test, y_train, y_test = train_test_split(q_df.div(4), df_LENs.loc[:, "EW_vlow":"RI_vhigh"], df_LENs.loc[:, "QOL_vlow"], test_size=0.2, random_state=42)


x_traint = torch.tensor(x_train.iloc[:, columns].values, dtype=torch.float)
x_testt = torch.tensor(x_test.iloc[:, columns].values, dtype=torch.float)
c_traint = torch.tensor(c_train.values, dtype=torch.float)
c_testt = torch.tensor(c_test.values, dtype=torch.float)
y_traint = torch.tensor(y_train.values, dtype=torch.float)
y_testt = torch.tensor(y_test.values, dtype=torch.float)

concept_encoder = torch.nn.Sequential(
    torch.nn.Linear(x_train.iloc[:, columns].shape[1], 2048),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(2048, 512),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(512, 120),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(120, c_traint.shape[1]),
    torch.nn.Sigmoid()
)

task_predictor = torch.nn.Sequential(
    te.nn.EntropyLinear(c_traint.shape[1], 128, n_classes=1),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(128, 8),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(8, 1)
)

model = torch.nn.Sequential(concept_encoder, task_predictor)
```

```python
#To load the model if already saved
model.load_state_dict(torch.load("drive/MyDrive/23QLensVlow2.pt"))
```

```python
optimizer = torch.optim.AdamW(model.parameters(), lr=0.01)
loss_form = torch.nn.BCEWithLogitsLoss()
loss_form2 = torch.nn.BCELoss()
```

```python
model.train()
for epoch in range(1000):
    if epoch % 100 == 1:
      print(epoch)
      c_predt = concept_encoder.forward(x_testt)
      y_predt = task_predictor(c_predt).reshape(-1)

      concept_accuracy = accuracy_score(c_testt > 0.5, c_predt > 0.5)
      task_accuracy = accuracy_score(y_testt, y_predt > 0.5)

      print(concept_accuracy)
      print(task_accuracy)

    optimizer.zero_grad()

    c_pred = concept_encoder(x_traint)
    y_pred = task_predictor(c_pred)

    loss_predictor = loss_form2(c_pred, c_traint) + 0.2*loss_form(y_pred.reshape(-1), y_traint) + 0.0001 * te.nn.functional.entropy_logic_loss(task_predictor)
    loss_predictor.backward()

    optimizer.step()
```

```python
from sklearn.metrics import accuracy_score

c_pred = concept_encoder.forward(x_testt)
y_pred = task_predictor(c_pred)

concept_accuracy = accuracy_score(c_test, c_pred > 0.5)
task_accuracy = accuracy_score(y_testt, y_pred.reshape(-1) > 0.5)

print(concept_accuracy)
print(task_accuracy)
```

```python
#To save the model in order to not train it again
#torch.save(model.state_dict(), "drive/MyDrive/23QLensVlow2.pt")
```

```python
global_explanations, local_explanations = entropy.explain_classes(task_predictor, concept_encoder.forward(x_traint), y_traint.reshape(-1, 1), c_threshold=0.5, y_threshold=0.5, max_minterm_complexity=10)
global_explanations
```

```python
global_explanations, local_explanations = entropy.explain_classes(task_predictor, concept_encoder.forward(x_traint), y_traint.reshape(-1, 1), c_threshold=0.5, y_threshold=0.5, max_minterm_complexity=10, simplify=True)
global_explanations
```
